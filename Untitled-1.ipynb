```python
import pandas as pd
import matplotlib.pyplot as plt
import missingno as msno
import plotly.express as px
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from nltk.corpus import stopwords
import string
import re
nltk.download('stopwords')
nltk.download('vader_lexicon')

stemmer = nltk.SnowballStemmer("english")
import warnings
warnings.filterwarnings ('ignore')
```

```python
#check missing ratio
df_na = (df.isnull().sum() / len(df)) * 100
df_na = df_na.drop(df_na[df_na == 0].index).sort_values(ascending=False)[:30]
missing_data = pd.DataFrame({'Missing Ratio' :df_na})
missing_data.head(20)
```

```python
def clean_data(df):
    df = df.dropna()
    df = df.drop(columns=['Image_Links'])
    df = df.rename(columns={'name': 'Name'})
    df.insert(2, "City", df["location"].str.split(",").str[0])
    df['Date'] = df['Date'].str.replace("Reviewed", "", case=False, regex=False)
    df = df.astype({'Date': 'datetime64[ns]'})
    df = df.astype({'Rating': 'int64'})
    df = df.drop(columns=['location'])

    return df

df_clean = clean_data(df.copy())
```

```python
df_clean['Month'] = df_clean['Date'].dt.month
df_clean['Year'] = df_clean['Date'].dt.year
df_clean['DayOfWeek'] = df_clean['Date'].dt.dayofweek
```

```python
#Clean Review
stopword=set(stopwords.words('english'))
def clean(text):
    text = str(text).lower()
    text = re.sub('\[.*?\]', '', text)
    text = re.sub('https?://\S+|www\.\S+', '', text)
    text = re.sub('<.*?>+', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\n', '', text)
    text = re.sub('\w*\d\w*', '', text)
    text = [word for word in text.split(' ') if word not in stopword]
    text=" ".join(text)
    text = [stemmer.stem(word) for word in text.split(' ')]
    text=" ".join(text)
    return text
df_clean["Review"] = df_clean["Review"].apply(clean)
```

```python
# Calculate the count of each rating
ratings = df_clean["Rating"].value_counts().reset_index()
ratings.columns = ['Rating', 'Count']

# Create a pie chart
fig = px.pie(
    ratings,
    values='Count',
    names='Rating',
    title='Rating Distribution',
    hole=0.5,  # Adjust the size of the hole in the pie chart
)

# Customize hovertemplate to include slash between Rating and Count
fig.update_traces(
    textinfo='percent+label',
    hovertemplate='%{label}: %{percent}<br>Count: %{value}',
)

fig.show()
```

```python
categorical_columns = ['Name', 'City']

for column in categorical_columns:
    # Get the top 15 categories by frequency
    top_10_categories = df_clean[column].value_counts().head(10)
    
    # Create a DataFrame for the top 15 categories and their frequencies
    top_10_df = pd.DataFrame({'Category': top_10_categories.index, 'Frequency': top_10_categories.values})
    
    fig = px.bar(
        top_10_df,
        x='Category',
        y='Frequency',
        title=f'Top 10 Categories in {column}',
        labels={'Category': column, 'Frequency': 'Frequency'},
        color='Category',
        color_discrete_sequence=px.colors.qualitative.Set3  # Customize the color scheme
    )
    
    fig.update_layout(
        xaxis=dict(title=column),
        yaxis=dict(title='Frequency'),
        xaxis_tickangle=-90,  
        showlegend=False  
    )
    
    fig.show()
```

```python
day_names = ['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday']
annual_rating = df_clean.groupby(['DayOfWeek'])['Rating'].sum().reset_index()

# Create a bar chart
fig = px.bar(
    annual_rating,
    x='DayOfWeek',
    y='Rating',
    labels={'DayOfWeek': 'DayOfWeek', 'Rating': 'Total Rating'},
    title='Total Rating Count Per DayOfWeek'
)

# Customize the x-axis ticks to display day names
fig.update_xaxes(
    tickmode='array',
    tickvals=annual_rating['DayOfWeek'].unique(),
    ticktext=day_names
)

fig.show()
```

```python
annual_rating = df_clean.groupby(['Month'])['Rating'].sum().reset_index()

# Create a line chart
fig = px.line(
    annual_rating,
    x='Month',
    y='Rating',
    labels={'Month': 'Month', 'Rating': 'Total Rating'},
    title='Total Rating Count Per Month'
)

fig.show()
```

```python
annual_rating = df_clean.groupby(['Year'])['Rating'].sum().reset_index()

fig = px.line(
    annual_rating,
    x='Year',
    y='Rating',
    labels={'Year': 'Year', 'Rating': 'Total Rating'},
    title='Total Rating Count Per Year'
)

fig.show()
```

```python
text = " ".join(i for i in df_clean.Review)
stopwords = set(STOPWORDS)
wordcloud = WordCloud(stopwords=stopwords, background_color="black").generate(text)
plt.figure( figsize=(15,10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
sentiments = SentimentIntensityAnalyzer()
df_clean["Positive"] = [sentiments.polarity_scores(i)["pos"] for i in df_clean["Review"]]
df_clean["Negative"] = [sentiments.polarity_scores(i)["neg"] for i in df_clean["Review"]]
df_clean["Neutral"] = [sentiments.polarity_scores(i)["neu"] for i in df_clean["Review"]]
df_clean = df_clean[["Review", "Positive", "Negative", "Neutral"]]
print(df_clean.head())
```

```python
positive =' '.join([i for i in df_clean['Review'][df_clean['Positive'] > df_clean["Negative"]]])
stopwords = set(STOPWORDS)
wordcloud = WordCloud(stopwords=stopwords, background_color="black").generate(positive)
plt.figure( figsize=(15,10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
